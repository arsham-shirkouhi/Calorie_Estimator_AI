````python
import os
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path

print('Libraries loaded successfully.')
````

````python
# Update this path to where you downloaded Food-101
DATA_DIR = Path('food-101/images')

# Get all class names and image counts
classes = sorted([d.name for d in DATA_DIR.iterdir() if d.is_dir()])
counts  = {c: len(list((DATA_DIR / c).glob('*.jpg'))) for c in classes}

print(f'Total classes   : {len(classes)}')
print(f'Total images    : {sum(counts.values())}')
print(f'Images per class: {list(counts.values())[0]} (uniform)')
````

````python
df_counts = pd.DataFrame({'food': list(counts.keys()), 'count': list(counts.values())})
df_counts = df_counts.sort_values('count', ascending=False)

fig, ax = plt.subplots(figsize=(18, 5))
ax.bar(df_counts['food'], df_counts['count'], color='steelblue', edgecolor='white', linewidth=0.3)
ax.set_title('Image Count per Food Category (Food-101)', fontsize=14, fontweight='bold')
ax.set_xlabel('Food Category')
ax.set_ylabel('Number of Images')
ax.set_xticks([])
plt.tight_layout()
plt.savefig('class_distribution.png', dpi=150)
plt.show()
print('Dataset is perfectly balanced — 1,000 images per class.')
````

````python
sample_classes = random.sample(classes, 12)

fig, axes = plt.subplots(3, 4, figsize=(14, 10))
for ax, cls in zip(axes.flatten(), sample_classes):
    img_path = random.choice(list((DATA_DIR / cls).glob('*.jpg')))
    ax.imshow(mpimg.imread(img_path))
    ax.set_title(cls.replace('_', ' ').title(), fontsize=9)
    ax.axis('off')

plt.suptitle('Random Sample Images from Food-101', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('sample_images.png', dpi=150)
plt.show()
````

````python
db = pd.read_csv('../data/nutrition_database.csv')
print(f'Database shape: {db.shape}')
db.head(10)
````

````python
print('Summary statistics (per 100g):')
db[['calories','protein','carbs','fat','fiber','sugar','sodium']].describe().round(1)
````

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

top10    = db.nlargest(10, 'calories')
bottom10 = db.nsmallest(10, 'calories')

axes[0].barh(top10['food'].str.replace('_',' ').str.title(), top10['calories'], color='tomato')
axes[0].set_title('Top 10 Highest Calorie Foods', fontweight='bold')
axes[0].set_xlabel('Calories (per 100g)')
axes[0].invert_yaxis()

axes[1].barh(bottom10['food'].str.replace('_',' ').str.title(), bottom10['calories'], color='mediumseagreen')
axes[1].set_title('Top 10 Lowest Calorie Foods', fontweight='bold')
axes[1].set_xlabel('Calories (per 100g)')
axes[1].invert_yaxis()

plt.tight_layout()
plt.savefig('calorie_distribution.png', dpi=150)
plt.show()
````

````python
avg = db[['protein','carbs','fat']].mean()

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(avg, labels=['Protein','Carbs','Fat'],
       autopct='%1.1f%%', colors=['#4C9BE8','#F4A261','#E76F51'],
       startangle=140, textprops={'fontsize': 12})
ax.set_title('Average Macronutrient Distribution\nacross Food-101 Categories', fontweight='bold')
plt.tight_layout()
plt.savefig('macronutrient_distribution.png', dpi=150)
plt.show()
```

---

Dokuzuncu hücre — bu **Markdown hücresi** olacak, "**+ Markdown**" seç:
```
## Key Observations

- Food-101 is a **perfectly balanced dataset** with exactly 1,000 images per class across 101 categories.
- The nutritional database covers all 101 categories with 7 nutritional attributes per food item (calories, protein, carbs, fat, fiber, sugar, sodium).
- Calorie content varies significantly — from light options like **miso soup (~40 kcal/100g)** to dense options like **foie gras (~462 kcal/100g)**.
- On average, carbohydrates make up the largest macronutrient share across all food categories.
- The balanced class distribution means **no oversampling or undersampling** will be needed during training.
````
